```python
import pandas as pd
import requests

API_KEY = "4d4ca9b81c5b43cdb92f90cc22e98042"  #API key
BASE_URL = "https://newsapi.org/v2/everything"

# Load tracked companies from Excel
companies = pd.read_excel("data/cleaned_companies.xlsx")["Company"]

all_news = []

for company in companies:
    params = {
        "q": company,
        "apiKey": API_KEY,
        "language": "en",
        "pageSize": 3,  # top 3 headlines per company
    }
    response = requests.get(BASE_URL, params=params)
    articles = response.json().get("articles", [])

    for art in articles:
        all_news.append({
            "Company": company,
            "Title": art["title"],
            "Source": art["source"]["name"],
            "URL": art["url"]
        })

# Save to Excel
df_news = pd.DataFrame(all_news)

print("Dataframe with news:")
print(df_news)

df_news.to_excel("data/company_news.xlsx", index=False)

print("✅ Company news digest saved to data/company_news.xlsx")
```
